```python
import pandas as pd
```

```python
# Charger les deux fichiers CSV
df_1 = pd.read_csv('data/silver_cleaned_I.csv')
df_2 = pd.read_csv('data/silver_cleaned_II.csv')

# ConcatÃ©ner les deux DataFrames
df = pd.concat([df_1, df_2], ignore_index=True)
```

```python
aggregations = {
    'Recency' : 'first',
    'Invoice' : 'count', # = Order count -> frequency
    'Total' : 'sum', # -> Monetary
    'Quantity' : 'sum', # = Total Quantity
    'Price' : 'mean',
    'StockCode' : 'count',
    'Percent_cancelled' : 'first',
    'Average_Quantity' : 'first',
    'Order_Quantity' : 'mean', #Average_order_quantity
    'Average_Price' : 'first',
    'Average_Basket' : 'first',
    'Country' : 'first',
    'Flag' : 'first',
}
```

```python
df_grouped = df.groupby('Customer ID').agg(aggregations)
```

```python
df_grouped = df_grouped.rename(columns={
    'Invoice':'Frequency',
    'Total':'Monetary',
    'Average_basket':'Ã˜_basket',
    'Average_Quantity' : 'Ã˜_Qty',
    'Order_quantity' : 'Ã˜_InvoiceQty',
    'StockCode':'âˆƒ_items', #âˆƒ = uniqueness
    'Price' : 'Ã˜_Price',
    'Cancelled_order_count' : 'n_Cancelled',
    'Percent_cancelled' : '%_Cancelled',
    })
```
